```python
import matplotlib.pylab as plt
import numpy as np
```

```python
x = np.linspace(0.5,4.0,1000)
```

```python
k1 = [-0.5641,-0.5459,-0.5373]
k2 = [0.0974,0.0908,0.0884]
labels = ['surf gen','surf fit 0','surf fit 1']

for a,b,label in zip(k1,k2,labels):
    y = 1.0 + a*x + b*x*x

    plt.plot(x,y,label=label)
    
y = np.exp(-0.520*x)
plt.plot(x,y,linewidth=2,label='neutrons')

y = np.exp(-0.00001*x)
plt.plot(x,y,linewidth=2,label='gammas')

plt.ylim(0,1.2)
plt.legend()
```
